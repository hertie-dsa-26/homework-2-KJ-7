# Exercises

## Exercise 1

Create a function `is_even(x)` that returns `True` if `x` is even and `False` otherwise.

## Exercise 2

Write a function that takes an integer $x$ > 1 as an input and returns the sum of all integers 1 to $x$, including $x$.

## Exercise 3

Write a function that checks if an integer is a multiple of another integer.

## Exercise 4

Write a function that takes an integer $x$ > 1 as an input and returns the sum of all integers 1 to $x$, including $x$, where $x$ is a multiple of 2, 3, or 5.

## Exercise 5

Leap years contain an additional day, February 29, to keep the calendar year in sync with the astronomical year. As per [Wikipedia](https://en.wikipedia.org/wiki/Leap_year#Gregorian_calendar):

> Every year that is exactly divisible by four is a leap year, except for years that are exactly divisible by 100, but these centurial years are leap years if they are exactly divisible by 400.

Implement a function that takes `year` and returns `True` if `year` is a leap year and `False` otherwise.

## Exercise 6

Implement a function that takes a list of years and returns a count of leap years. Use this function to calculate the number of leap years from 0 to 2026. How out of sync would the calendar and astronomical date be without leap years?

## Exercise 7

The Fibonacci sequence came up in lecture when Drew recommend using its values for story/experience points. The Fibonacci sequence is a sequence of numbers where each element is the sum of the two elements that precede it. The Fibonacci numbers are $F_0 = 0$, $F_1 = 1$, and $F_n = F_{n - 1} + F_{n - 2}$ for $n > 1$.

Implement a function that returns the $n$-th Fibonacci number ($F_n$).

## Exercise 8

Implement a function that returns a list of the Fibonacci numbers $F_0 \dots F_n$. You can reuse your function from exercise 6 or write a more efficient implementation.

## Exercise 9

Calculate the sum of the Fibonacci sequence up to $F_n$ for $n = [0, 1, 2 \dots 20]$.

## Exercise 10

Take the Fibonacci sequence up to $F_{50}$. Create separate lists for the even and odd elements. How many even elements are there? How many odd elements are there? What are the sums of the two lists?
